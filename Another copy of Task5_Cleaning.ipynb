```python
import pandas as pd
import numpy as np

df = pd.read_csv(
    "https://raw.githubusercontent.com/YBI-Foundation/Dataset/main/Students%20Performance.csv"
)

df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
import numpy as np

# Numeric columns → mean
numeric_cols = df.select_dtypes(include=np.number).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())

# Categorical columns → mode
categorical_cols = df.select_dtypes(include='object').columns
for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
before = df.shape[0]

df = df.drop_duplicates()

after = df.shape[0]

print("Rows before:", before)
print("Rows after:", after)
```

```python
df['gender'] = df['gender'].astype('category')
df.dtypes
```

```python
df['average_score'] = (
    df['math score'] +
    df['reading score'] +
    df['writing score']
) / 3

df.head()
```

```python
def performance_category(score):
    if score >= 75:
        return "High"
    elif score >= 50:
        return "Medium"
    else:
        return "Low"

df['performance_category'] = df['average_score'].apply(performance_category)

df[['average_score', 'performance_category']].head()
```

```python
df.to_csv("cleaned_data.csv", index=False)
```

### Data Cleaning & Preprocessing Notes

1. The dataset was loaded using pandas `read_csv()` to begin the data analysis process.
2. Initial exploration was done using `head()` and `info()` to understand the dataset structure and datatypes.
3. Missing values were checked using `isnull().sum()` to identify columns requiring cleaning.
4. Although no missing values were found, standard imputation techniques were applied:
   - Mean imputation for numerical columns
   - Mode imputation for categorical columns
5. Duplicate records were checked and removed using `drop_duplicates()` to ensure data quality.
6. Datatype conversion was performed using `astype()` to optimize memory usage and improve analysis accuracy.
7. A new feature `average_score` was created by combining math, reading, and writing scores.
8. Another derived column `performance_category` was created to classify students based on their average score.
9. The cleaned and transformed dataset was saved as a CSV file for further analysis and reporting.
